```
#Aula 1.6 2020/09/21
#Importando as bibliotecas para iniciar a utilização do KMeans
import numpy as np
from matplotlib import pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

# Passo a passo do Kmeans
Criação do conjunto de dados

```
dataset = np.array([[1,4],[2,2],[2,5],[3,3],[3,5],[4,7],[5,6],[6,4],[6,7],[7,6],[7,9],[8,7],[8,9],[9,4],[9,8]])
plt.scatter(dataset[:,0],dataset[:,1])
plt.xlim(0,10)
plt.ylim(0,10)
plt.grid()
```

```
#Begin K-mean Algorithm
kmeans = KMeans(n_clusters=3, #number of clusters
                init='k-means++', #method of initialization
                n_init=10,
                max_iter=300) # iterações
pred_y = kmeans.fit_predict(dataset)
#End K-mean Algorithm

plt.scatter(dataset[:,0],dataset[:,1]) #plotar um gráfico de dispersão
plt.xlim(0,10)
plt.ylim(0,10)
plt.grid()

plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],s=75,c='red')
plt.show()
```

```
plt.scatter(dataset[:,0],dataset[:,1],c=pred_y)
plt.xlim(0,10)
plt.ylim(0,10)
plt.grid()

plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],s=70, c='red')
plt.show()
```

```
#Generate random data set
X,y = make_blobs(n_samples=500,centers=50,random_state=0)
plt.scatter(X[:,0],X[:,1])
```

```
  #choosing the method to elect the number of clusters
  wcss =[]
  for i in range(1,20):
    kmeans = KMeans(n_clusters=i,init='k-means++',max_iter=300, n_init=10)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,20),wcss)
plt.title("Elbow Method")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.show()
```

```
#analysing the graph above, we choose the best number of cluets which is minimum clusters by optimal variation (I need to think deeper on this topic)
kmeans = KMeans(n_clusters=5, init='k-means++',max_iter=300, n_init=10)
pred_y = kmeans.fit_predict(X)

plt.scatter(X[:,0],X[:,1],c=pred_y)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=70,c='red')
plt.show()
```
